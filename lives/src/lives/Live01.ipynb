```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import polars as pl
```

```python
students = pl.read_csv("./data/StudentsPerformance.csv", separator=";")
students.head()
```

```python
students.select(["math score", "reading score", "writing score"]).describe()
```

```python
sns.heatmap(students_corr)
```
